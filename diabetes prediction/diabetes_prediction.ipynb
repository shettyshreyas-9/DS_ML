Import the dependencies

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Separating Features & Target

```python
X = df.drop(columns=['Outcome'])
X.head()
```

```python
Y = df['Outcome']
Y.head()
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(X)
```

```python
X_transf = scaler.fit_transform(X)
X_transf
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_transf, Y, test_size=0.2,stratify=Y, random_state= 2)
```

```python
print(X_transf.shape,X_train.shape, X_test.shape)
```

Training the Model

```python
clf = svm.SVC(kernel='linear')
```

```python
# training the svm classifier

clf.fit(X_train,Y_train)
```

Model Prediction & Evaluation

```python
Y_pred= clf.predict(X_test)
```

```python
accuracy_score(Y_pred,Y_test)
```

Making a Predictive system

```python
ip_data = (7,196,90,0,0,39.8,0.451,41)

# change the dtype to ndarray  & reshaping it
ip_data_as_array= np.array(ip_data).reshape(1,-1)

# standardize the ip_data
std_data= scaler.transform(ip_data_as_array)
# std_data

predictor= clf.predict(std_data)

# print(predictor)

if predictor[0] == 0:
  print('Diabetic')
else:
  print('Non-diabetic')
```
